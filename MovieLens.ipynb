## Predictive Model

```python
from sklearn.model_selection import cross_val_score as cvs
```

```python
x = data5k[['Age', 'MovieID']]
y = data5k['Rating']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
scaler = preprocessing.StandardScaler().fit(x_train)  #data processing
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
len(y_pred)
y_pred1 = knn.predict_proba(x_test)
y_pred1
```

```python
print(classification_report(y_test,y_pred))
```

## Grid Search CV and Random Search CV

```python
from sklearn.model_selection import GridSearchCV as gs, RandomizedSearchCV as rs
```

```python
params = {'p' : [1,2,3], 'n_neighbors': range(1,30,2), 'weights': ['uniform','distance'] }
```

```python
grid = gs(estimator = knn, param_grid = params)
grid.fit(x_train,y_train)
print('Best Score :',grid.best_score_)
print('Neighbors:', grid.best_estimator_.n_neighbors)
```

```python
grid = gs(estimator = knn, param_grid = params)
grid.fit(x_test,y_test)
print('Best Score :',grid.best_score_)
print('Neighbors:', grid.best_estimator_.n_neighbors)
```

```python
rsearch = rs(estimator = knn, param_distributions = params,cv=4,n_iter=8,random_state=5)
rsearch.fit(x_train,y_train)
print('Best Score :',rsearch.best_score_)
print('Neighbors:', rsearch.best_estimator_.n_neighbors)
```

```python
rsearch = rs(estimator = knn, param_distributions = params,cv=4,n_iter=8,random_state=5)
rsearch.fit(x_test,y_test)
print('Best Score :',rsearch.best_score_)
print('Neighbors:', rsearch.best_estimator_.n_neighbors)
```

### From the above 'Best Score', we can conclude that our prediction model accurate and has nill underfitting or overfitting problems.
